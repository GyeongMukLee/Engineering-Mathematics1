# Numpy를 사용하여 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

y = 3x 그래프는 아래와 같이 그릴 수 있다.

```python
graph1_x = np.linspace(0, 3, 30)
graph1_y = graph1_x * 3

plt.plot(graph1_x, graph1_y)
```

y = x^2의 그래프는 아래와 같이 그릴 수 있다.

```python
graph2_x=np.linspace(-1, 1, 30)
graph2_y = graph2_x ** 2

plt.plot(graph2_x, graph2_y)
```

## 파일에서 데이터를 받아와 그래프 그리기

txt 파일에서 데이터를 가져와 저장하는 numpy 함수

```python
txtdata = np.loadtxt("./Chapter14/populations.txt")
print(txtdata)
```

```python
data_year, data_hares, data_lynxes, data_carrots = txtdata.T

print(data_year)
print(data_hares)
print(data_lynxes)
print(data_carrots)
```

```python
plt.axes([0.2, 0.1, 0.5, 0.8])
plt.plot(data_year, data_hares, data_year, data_lynxes, data_year, data_carrots)
plt.legend(("Hare", "Lynx", "Carrot"), loc=(1.05, 0.5))
```
