# 튜플


## 튜플이란
튜플은 순서가 있고, 변하지 않는 자료형이다.

```python
week = ("Mon", "Tue", "Wed", "Thr", "Fri", "Sat", "Sun")
```

괄호를 생략해도 튜플이 될 수 있다.

```python
istuple = 10, 20, 30, 40, 50
print(istuple)
print(type(istuple))
```

## 튜플의 인덱스
튜플도 다른 시퀀스 자료형과 마찬가지로 인덱스를 지정할 수 있다.

```python
print(week[3])
print(week[-2])
print(week[::2])
```

## 튜플과 함수
함수에서 값을 여러 개 반환할 때 튜플을 사용한다.

```python
def squareAndCube(x):
    return x**2, x**3


print(squareAndCube(3))
print(type(squareAndCube(3)))
```

## 튜플 값의 변경
튜플은 원소의 값을 변경할 수 없다.

```python
try:
    week[0] = "Sun"
except TypeError as e:
    print(e)
```

## 길이가 1인 튜플 만드는 방법

```python
single=(10,)
print(single)
print(type(single))
```

## 튜플의 메소드
튜플은 대표적으로 두 개의 메소드를 가지고 있다.

### tuple.count()
count 메소드는 튜플에서 괄호 안에 입력한 객체가 몇 개 있는지 세는 함수이다.

```python
week.count("Sun")
```

### tuple.index()
index 메소드는 괄호 안에 입력한 객체가 몇 번째에 있는지 나타내는 함수이다.
같은 객체가 여러 개 있을 경우 맨 처음에 있는 객체를 대상으로 한다.

```python
week.index("Wed")
```
